```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern

radius = 1
n_points = 8 * radius
method1 = 'default'
method2 = 'ror'

image = cv2.imread('image3.png', cv2.IMREAD_GRAYSCALE)

lbp_default = local_binary_pattern(image, n_points, radius, method=method1)
lbp_ri = local_binary_pattern(image, n_points, radius, method=method2)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(lbp_default.ravel(), bins=int(lbp_default.max()+1), density=True)
plt.title("Original LBP Histogram")

plt.subplot(1, 2, 2)
plt.hist(lbp_ri.ravel(), bins=int(lbp_ri.max()+1), density=True)
plt.title("Rotated Invariant LBP Histogram")

plt.tight_layout()
plt.show()
```
