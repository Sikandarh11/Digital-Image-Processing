```python
img = cv2.imread("fig01.tif", 0)
def power_law(img, gamma):
    s = np.zeros_like(img, dtype=np.int32)
    for i in range(img.shape[0]): 
        for j in range(img.shape[1]): 
            s[i,j] = (np.power(img[i,j]/255,gamma))*255
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(img, cmap='gray')
    axes[0].set_title("Original")
    axes[1].imshow(s, cmap='gray')
    axes[1].set_title("Solved power Law")
 
```

```python
img = cv2.imread("fig03.tif", 0)
def log_law(img):
    s = np.zeros_like(img, dtype=np.int32)
    max_v = int(np.max(img))
    print(max_v)
    c=int(255/np.log10(1+max_v) )
    print(c)
    count = 0
    for i in range(img.shape[0]): 
        for j in range(img.shape[1]): 
            if img[i,j] == 0:
                s[i,j] = (c*np.log10(1+img[i,j]))
                #print(s[i,j])
                count+=1
            else:
                s[i,j] = (c*np.log10(img[i,j]))
                count+=1
    print(count)
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(img, cmap='gray')
    axes[0].set_title("Original ")
    axes[1].imshow(s, cmap='gray')
    axes[1].set_title("After Log Transform")
 
```

```python
img = cv2.imread("gradient.png", 0)
def slicing(img):
    s = img.copy()
    for i in range(img.shape[0]): 
        for j in range(img.shape[1]): 
            if img[i,j] >= 100 and img[i,j] <= 200:
                s[i,j] = 210
                
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(img, cmap='gray')
    axes[0].set_title("Original Binary")
    axes[1].imshow(s, cmap='gray')
    axes[1].set_title("After Slicing Transform")
```
